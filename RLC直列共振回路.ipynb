```python
import numpy as np
import matplotlib.pyplot as plt
!pip install japanize_matplotlib
import japanize_matplotlib
```

```python
# 回路素子の値
E = complex(10, 0) # [V]
R = 10             # [Ohm]
L = 10E-3          # [H]
C = 4E-6           # [F]

# 角周波数
omega = np.linspace(1, 1E+4, 200)

# 電流の絶対値
I_abs = E/np.sqrt(R**2 + (omega*L-1/(omega*C))**2)

# 電流の位相
I_arg = -np.arctan2(omega*L-1/(omega*C), R)


# グラフを描画
plt.rcParams["font.size"]=14
plt.figure(figsize=(6, 6))

plt.subplot(2, 1, 1)
plt.plot(omega, I_abs)
plt.xlabel("角周波数, $\omega$ [rad/s]")
plt.ylabel("電流の大きさ, |$I$| [A]")
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(omega, I_arg)
plt.xlabel("角周波数, $\omega$ [rad/s]")
plt.ylabel("電流の位相, arg($I$) [A]")
plt.yticks([-np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2])
plt.grid()
plt.tight_layout()
```

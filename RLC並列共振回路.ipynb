$\displaystyle \boldsymbol{J}=J_ee^{j0}=J_e$ [A]

$\displaystyle \boldsymbol{Y}=\frac{1}{R}+j\left(\omega C - \frac{1}{\omega L} \right)=\sqrt{\frac{1}{R^2}+\left(\omega C - \frac{1}{\omega L} \right)^2}e^{j\theta}$ [$\Omega$]

$\displaystyle \theta=\tan^{-1}\frac{\omega C-1/\omega L}{1/R}$ [rad]

$\displaystyle \boldsymbol{V}=\frac{\boldsymbol{J}}{\boldsymbol{Y}}=\frac{J_e}{\sqrt{\frac{1}{R^2}+\left(\omega C - \frac{1}{\omega L} \right)^2}}e^{-j\theta}$ [V]

$\displaystyle |\boldsymbol{V}|=\frac{J_e}{\sqrt{R^2 + \left(\omega L -\frac{1}{\omega C} \right)^2}}$ [A]

```python
# 回路素子の値
J = 10+0j       # [A]
R = 300         # [Ohm]
L = 10E-3       # [H]
C = 4E-6        # [F]

# 角周波数
omega = np.linspace(1, 1E+4, 200) # [rad/s]

# 電圧の絶対値
V_abs = abs(J)/np.sqrt((1/R)**2 + (omega*C-1/(omega*L))**2)

# 電圧の位相
V_arg = -np.arctan2(omega*C-1/(omega*L), 1/R)


# グラフを描画
plt.rcParams["font.size"]=14
plt.figure(figsize=(6, 6))

plt.subplot(2, 1, 1)
plt.plot(omega, V_abs)
plt.xlabel("角周波数, $\omega$ [rad/s]")
plt.ylabel("電圧の大きさ, |$V$| [A]")
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(omega, V_arg)
plt.xlabel("角周波数, $\omega$ [rad/s]")
plt.ylabel("電圧の位相, arg($V$) [A]")
plt.yticks([-np.pi/2, -np.pi/4, 0, np.pi/4, np.pi/2])
plt.grid()
plt.tight_layout()
```
